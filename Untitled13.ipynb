```python
import pandas as pd
df1=pd.read_csv('Sales_April_2019.csv', delimiter=',')
df2=pd.read_csv('Sales_August_2019.csv', delimiter=',')
df3=pd.read_csv('Sales_December_2019.csv', delimiter=',')
df4=pd.read_csv('Sales_February_2019.csv', delimiter=',')
df5=pd.read_csv('Sales_January_2019.csv', delimiter=',')
df6=pd.read_csv('Sales_July_2019.csv', delimiter=',')
df7=pd.read_csv('Sales_June_2019.csv', delimiter=',')
df8=pd.read_csv('Sales_November_2019.csv', delimiter=',')
df9=pd.read_csv('Sales_October_2019.csv', delimiter=',')
df10=pd.read_csv('Sales_September_2019.csv', delimiter=',')
df11=pd.read_csv('Sales_March_2019.csv', delimiter=',')
df12=pd.read_csv('Sales_May_2019.csv', delimiter=',')
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, ], axis=0, ignore_index=True)
```

```python
print(df.head())
```

```python
df['Order ID'] = pd.to_numeric(df['Order ID'], errors='coerce')
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce')
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce').astype(float)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M', errors='coerce')
```

```python
# Extrait le mois et l'ann√©e de la colonne "Order Date"
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Price Each'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Cr√©e un graphique des ventes mensuelles
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str), monthly_sales['Price Each'], marker='o')
plt.xlabel('Mois')
plt.ylabel('Ventes Mensuelles')
plt.title('Ventes Mensuelles au fil du Temps')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
